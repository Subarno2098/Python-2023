````python
import geopandas as gpd
from eodag import EODataAccessGateway
import folium
import glob
import rioxarray
from rioxarray import merge
import numpy as np

lf = gpd.read_file('../Data/vector/example/lower_frankonia.gpkg')
lf.explore()
````

````python
xn,yn,xx,yx = lf.total_bounds

dag = EODataAccessGateway()

products, total_count = dag.search(
   productType="cop-dem-glo-90",
   geom=(xn,yn,xx,yx)
)

print(total_count)
````

````python
workspace = '../Data/raster/example'
paths = dag.download_all(
    products,
    outputs_prefix=workspace,
)
````

````python
emap = folium.Map([48.5, 13], zoom_start=6)

layer = folium.features.GeoJson(
    data=products.as_geojson_object(),
    tooltip = folium.GeoJsonTooltip(fields=['title'])
).add_to(emap)

emap
````

````python
search_path = f'{workspace}/*/Copernicus*.tif'
items = glob.glob(search_path, recursive=True)
elements = []
for val in items:
    elements.append(rioxarray.open_rasterio(val))

merged = merge.merge_arrays(elements)
lf_raster = merged.rio.clip(lf.geometry.values, lf.crs)
lf_raster
````

````python
lf_raster.plot()
````

````python
elev = lf_raster.values

def hillshade(array, azimuth, angle_altitude): 
    x, y = np.gradient(array)
    slope = np.pi/2. - np.arctan(np.sqrt(x*x + y*y))
    aspect = np.arctan2(-x, y)
    azimuthrad = azimuth * np.pi / 180.
    altituderad = angle_altitude*np.pi / 180.
    shaded = np.sin(altituderad) * np.sin(slope)\
     + np.cos(altituderad) * np.cos(slope)\
     * np.cos(azimuthrad - aspect)
    return 255*(shaded + 1)/2

hill = hillshade(elev[0],315, 45)
````

````python
from matplotlib import colors
import matplotlib.pyplot as plt
from matplotlib import font_manager
from palettable.cartocolors.sequential import Magenta_7, PurpOr_5, Sunset_5, Teal_7, agGrnYl_5_r
from palettable.cmocean.sequential import Deep_14, Matter_20, Dense_20
from palettable.matplotlib import Magma_20_r

value_range = np.nanmax(elev) + abs(np.nanmin(elev))
topo = elev[0] + abs(np.nanmin(elev))

fig, ax = plt.subplots(figsize=(14,14))
plt.imshow(topo, cmap='terrain')
ax.imshow(hill, cmap="Greys", alpha=0.3, vmin=50)
ax.axis('off')
plt.figtext(0.055, 0.76, "Digital Surface Model of".upper(), fontsize=12,  ha="left")
plt.figtext(0.05, 0.73, "Lower Frankonia".upper(), fontsize=30,  ha="left")
plt.figtext(0.05, 0.1, "Data: Shuttle Radar Topography Mission  |  By: Steven Hill", fontsize=10)
plt.savefig("shaded-relief-lf.png", bbox_inches='tight', pad_inches=0.45)
plt.show()
````

````python
import ee 
from ee_plugin import Map 

elev = ee.Image('USGS/GMTED2010')

Map.setCenter(-121.069, 50.709, 6)

Map.addLayer(elev, {}, 'elev')

shade = ee.Terrain.hillshade(elev)
Map.addLayer(shade, {}, 'hillshade', False)

sea = elev.lte(0)
Map.addLayer(sea.mask(sea), {'palette':'000022'}, 'sea', False)

elevationPalette = ['006600', '002200', 'fff700', 'ab7634', 'c4d0ff', 'ffffff']
visParams = {'min': 1, 'max': 3000, 'palette': elevationPalette}

visualized = ee.ImageCollection([
  elev.mask(sea.Not()).visualize(**visParams),

  sea.mask(sea).visualize(**{'palette':'000022'})
]).mosaic()


Map.addLayer(visualized, {}, 'elev palette', False)

hsv = visualized.divide(255).rgbToHsv()
hs = hsv.select(0, 1)
v = shade.divide(255)
rgb = hs.addBands(v).hsvToRgb().multiply(255).byte()
Map.addLayer(rgb, {}, 'styled')
````
